```python
from google.colab import drive
drive.mount('/content/drive')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_json("/content/drive/MyDrive/Colab Notebooks/data.json")
data.head(5)
```

```python
data.shape
```

```python
ingredients = []

for array in tqdm(data.ingredients):
  for item in array:
    ingredients.append(item)
```

```python
ingredients[:5]
```

```python
unique_ingredients = []
for ingredient in ingredients:
  if ingredient not in unique_ingredients:
    unique_ingredients.append(ingredient)
len(unique_ingredients)
```

```python
N = len(data)
for ingredient in unique_ingredients:
  data[ingredient] = np.array([0]*N)
data.head(5)
```

```python
import pandas as pd
new_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/new_data.csv")
```

```python
new_data.head(5)
```

```python
x = new_data.drop(["Unnamed: 0", "id", "cuisine", "ingredients"], axis = 1)
```

```python
x.head()
```

```python
y = new_data.cuisine
y.head()
```

```python
y.hist(bins = 70, xrot = 45)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.75)
```

```python
del new_data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression(n_jobs = -1)

model.fit(x_train, y_train)
y_pred = model.predict(x_test)

accuracy_score(y_test, y_pred)
```

```python
y_pred[:5]
```

```python
from sklearn.tree import DecisionTreeClassifier

model_tree = DecisionTreeClassifier()


model_tree.fit(x_train, y_train)
y_pred_tree = model_tree.predict(x_test)

accuracy_score(y_test, y_pred_tree)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_tree = RandomForestClassifier()


model_tree.fit(x_train, y_train)
y_pred_tree = model_tree.predict(x_test)

accuracy_score(y_test, y_pred_tree)
```
